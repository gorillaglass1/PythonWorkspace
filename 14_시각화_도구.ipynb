```python
import pandas as pd
from os import path

def getData():
  return pd.read_csv(path.join("traffic.csv"), encoding='euc-kr')

df = getData()
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

df = getData()
pclass = df.구분1.value_counts()

label = ['location : %s'%pclass for pclass in pclass.index]
label

fig, ax = plt.subplots()
ax.pie(pclass, labels=label, autopct='%1.1f%%', shadow=True)
ax.axis('equal')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = getData()

plt.figure(figsize=(12, 6)) # Adjust figure size for better readability
ax = sns.countplot(x="구분1", y="평일", data=df)
ax = sns.barplot(x="구분1", y="평일", data=df)
ax.set(xlabel="지역", ylabel="평일 사용금액")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = getData()

plt.figure(figsize=(12, 6)) # Adjust figure size for better readability
ax = sns.barplot(x="구분1", y="평일", data=df)
ax.set(xlabel="지역", ylabel="평일 사용금액")
plt.title("지역별 평일 사용금액") # Add a title to the plot
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability if they overlap
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
